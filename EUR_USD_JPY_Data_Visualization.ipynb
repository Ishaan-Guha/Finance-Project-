```python
df = pd.read_csv("USD EURO JPY Clean Data.csv")

print(df.columns)

df.columns = df.columns.str.strip()

df = df.rename(columns={"Dates":"Date"})

print(df)

df["Date"] = df["Date"].astype(str)

df["Date"] = pd.to_datetime(df["Date"], dayfirst = True, errors = "coerce")

df = df.dropna(subset=["Date"])

df = df.sort_values("Date")
```

```python
plt.plot(df["Date"], df["Daily Close"])
plt.show()
```

```python
plt.plot(df["Date"], df["Daily Returns"])
plt.axhline(0)
plt.show()
```

```python
#FX Price Trend

plt.figure(figsize=(12,5))
plt.plot(df["Date"], df["Daily Close"])
plt.title("FX Exchange Rate Trend")
plt.xlabel("Date")
plt.ylabel("Exchange Rate")
plt.show()
```

```python
#Daily Returns

plt.figure(figsize=(12,5))
plt.plot(df["Date"], df["Daily Returns"])
plt.axhline(0)
plt.title("Daily FX Returns")
plt.xlabel("Date")
plt.ylabel("Daily Returns")
plt.show()
```
